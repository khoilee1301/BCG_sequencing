```python
#function to check and find base pair differences between the reference sequences and inserted Pasteur sequences.
def find_identical_locations(seq1, seq2):
    locations = []
    seq1_mismatch = []
    seq2_mismatch = []
    # Check if sequences are the same length first for a more complete check
    if len(seq1) != len(seq2):
         print("Sequences have different lengths, they cannot be identical.")
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            locations.append(i)
            seq1_mismatch.append(seq1[i])
            seq2_mismatch.append(seq2[i])
    return locations, seq1_mismatch, seq2_mismatch
```
